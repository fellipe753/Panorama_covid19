COVID-19 √© uma doen√ßa infecciosa causada por uma cepa de coronav√≠rus (SARS-CoV-2) que ainda n√£o havia sido identificada pelos cientistas. 

Os primeiros casos surgiram no fim do ano de 2019 em Wuhan na China. No fim de janeiro de 2020 a OMS j√° declarava o surto como uma Emerg√™ncia de S√°ude P√∫blica de Import√¢ncia Internacional.

As principais formas de transmiss√£o do v√≠rus √© por contato, transmiss√£o por got√≠culas e transmiss√£o por aerossol (got√≠culas respirat√≥rias minusculas).

Com a pandemia declarada, todo o mundo teve que mudar a forma de viver e em alguns casos foram aplicadas longos lockdowns, em outros tentaram seguir a rotina como antes, mas em todos os casos tivemos impactos na vida das pessoas e nas economias dos pa√≠ses.

Com o surgimento das vacinas, a letalidade da doen√ßa come√ßou a cair e as interna√ß√µes graves reduziram bastante, sendo a grande maioria das interna√ß√µes em pessoas que n√£o haviam sido vacinadas ou n√£o haviam completado a imuniza√ß√£o.

Mesmo com todos os esfor√ßos, novas variantes (muta√ß√µes do v√≠rus) surgiram, fazendo que mesmo com a vacina√ß√£o tiv√©ssemos uma explos√£o de novos casos, entretando com uma taxa de letalidade menor do que o registrado antes da vacina√ß√£o.

Desta forma, o projeto tem como objetivo ilustrar n√∫meros da pandemia no mundo, nos pa√≠ses e no Brasil. Os dados foram obtidos via [Our World in Data](https://github.com/owid/covid-19-data/tree/master/public/data) (OWID) e s√£o atualizados diariamente no github.

#1 - Import dos dados

Os dados est√£o dispon√≠veis em https://github.com/owid/covid-19-data/tree/master/public/data

```python
#@title
import pandas as pd
import matplotlib.pyplot as plt
import datetime
from datetime import datetime


%matplotlib inline
```

```python
#@title
#importando o csv direto do git
df = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv')#,index_col="date", parse_dates=True

#transformando a coluna date em datetime
df.date = pd.to_datetime(df.date, format='%Y-%m-%d')
#passando a coluna date como index
df.set_index('date', inplace=True)
```

```python
#@title
df.sample(5)
```

```python
#@title
print(f'Data inicial dos registros: {df.index.min()}')
print(f'Data final dos registros: {df.index.max()}')
```

```python
#@title
print(f'Total de vari√°veis no dataset: {len(df.columns)}')
print(f'Total de registro no dataset: {len(df)}\n')
```

```python
#@title
#ajustando a √∫ltima data para plot nos gr√°ficos
last_date = str(df.index[-1]).split(" ")
last_date =datetime.strptime(last_date[0], '%Y-%m-%d')
last_date = last_date.strftime('%d-%m-%Y')
```

#2 - Tratamento de Dados Faltantes

Colunas com alta ocorr√™ncia de dados em branco (>20%) ser√£o exclu√≠das do dataset.
Alguns destes dados se deve a dificuldade dos pa√≠ses e regi√µes de coletar e reportar os dados, por isto a defasagem.

Al√©m disso, em alguns casos se d√° pelo registro ter come√ßado a mais de um ano ap√≥s o in√≠cio da pandemia, como por exemplo dados de vacina√ß√£o. Estes dados em espec√≠ficos s√£o usados a partir do *dataframe* original.

```python
#@title
#pd.set_option('display.max_rows', None) # mostrar todas vari√°veis e n√£o s√≥ uma parte
(df.isnull().sum() / len(df) * 100).sort_values(ascending=False)
```

```python
#@title
print(f'Total de vari√°veis antes da remo√ß√£o de vari√°veis: {len(df.columns)}')

variaveis_incompletas = (df.isnull().sum() / len(df) * 100)
variaveis_incompletas = variaveis_incompletas[variaveis_incompletas>20]
df_clean = df.drop(columns=variaveis_incompletas.index)

print(f'Total de vari√°veis ap√≥s a remo√ß√£o de vari√°veis com 20% ou mais de nulos: {len(df_clean.columns)}\n')
df_clean.tail()
```

#3 - An√°lise Explorat√≥ria Mundo

Abaixo √© ilustrada uma an√°lise dos dados registrados para o Mundo.
Todos os dados s√£o resultados de um somat√≥rio dos dados di√°rios de todos pa√≠ses e regi√µes presentes no dataset.

```python
#@title
df_world = df_clean.copy()
df_world = df_world.loc[df_world['location'] == 'World']
# print(f' A rela√ß√£o entre casos e mortes (taxa de letalidade) causadas pela COVID-19 no mundo √© de: {round(df_world["total_deaths"][-1]/(df_world["total_cases"][-1])*100,2)}% na data {last_date}.')
```

##3.2 - Novos casos confirmados de COVID-19 no mundo

## 3.6 - Taxa de letalidade antes e depois da vacina√ß√£o

Foram utilizadas as vari√°veis `new_cases_smoothed_per_million` e `new_deaths_smoothed_per_million` das datas indicadas.

A partir destas vari√°veis, foi calculado a taxa de letalidade naquele dia (com m√©dia m√≥vel de 7 dias).

```python
#@title
df_world_vacinacao = df.loc[df['location'] == 'World']
primeira_vacina_world = df_world_vacinacao.loc[df_world_vacinacao['total_vaccinations'] >= 0.0].head(1)

data_primeira_vacina_world = primeira_vacina_world.index[0]
data_primeira_vacina_world = str(data_primeira_vacina_world).split(" ")
data_primeira_vacina_world = datetime.strptime(data_primeira_vacina_world[0], '%Y-%m-%d')
data_primeira_vacina_world = data_primeira_vacina_world.strftime('%d-%m-%Y')

tx_letalidade_world_antes_vacina = round(primeira_vacina_world["new_deaths_smoothed_per_million"][0] / primeira_vacina_world["new_cases_smoothed_per_million"][0] * 100,2)
tx_letalidade_world_atual = round(df_world_vacinacao["new_deaths_smoothed_per_million"][-1]/(df_world_vacinacao["new_cases_smoothed_per_million"][-1])*100,2)

print(f'A taxa de letalidade registrada na data da primeira vacina√ß√£o ({data_primeira_vacina_world}) era de {tx_letalidade_world_antes_vacina}%.')
print(f'A taxa de letalidade registrada na data atual ({last_date}) √© de {tx_letalidade_world_atual}%;')
```

## 5.5 - Letalidade antes e depois da vacina√ß√£o

Foram utilizadas as vari√°veis `new_cases_smoothed_per_million` e `new_deaths_smoothed_per_million` das datas indicadas.

A partir destas vari√°veis, foi calculado a taxa de letalidade naquele dia (com m√©dia m√≥vel de 7 dias).

```python
#@title
primeira_vacina_br = df_brasil.loc[df_brasil['total_vaccinations'] >= 0.0].head(1)

data_primeira_vacina_br = primeira_vacina_br.index[0]
data_primeira_vacina_br = str(data_primeira_vacina_br).split(" ")
data_primeira_vacina_br = datetime.strptime(data_primeira_vacina_br[0], '%Y-%m-%d')
data_primeira_vacina_br = data_primeira_vacina_br.strftime('%d-%m-%Y')

tx_letalidade_br_antes_vacina = round(primeira_vacina_br["new_deaths_smoothed_per_million"][0] / primeira_vacina_br["new_cases_smoothed_per_million"][0] * 100,2)
tx_letalidade_br_atual = round(df_brasil["new_deaths_smoothed_per_million"][-1]/(df_brasil["new_cases_smoothed_per_million"][-1])*100,2)

print(f'A taxa de letalidade registrada na data da primeira vacina√ß√£o ({data_primeira_vacina_br}) era de {tx_letalidade_br_antes_vacina}%.')
print(f'A taxa de letalidade registrada na data atual ({last_date}) √© de {tx_letalidade_br_atual}%.')
```
